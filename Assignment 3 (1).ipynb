```python
driver = webdriver.Chrome("chromedriver.exe")
#Q1.webpage
url = 'https://www.amazon.in./'  
driver.get(url)

#Enter "Data Analyst" in "Skill,Designations,Comapnies" field
search_job = driver.find_element_by_id('twotabsearchtextbox')
search_job.send_keys("guitar")


#Click the search button
search_btn = driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']")
search_btn.click()
```


In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product vertical has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
#fetching urls
URL = []
for i in range(0,3):
    lnk = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
    for j in lnk:
        URL.append(j.get_attribute('href'))
    nxt_page = driver.find_element_by_xpath('//li[@class="a-last"]')
    time.sleep(4)
    nxt_page.click()
    time.sleep(3)
```

#"Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. 
#In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
Brand= [] 
Product_title = []
Rating = []
No_ratings = []
Price = []
Return = []
Expected_delivery = []
Availability = []
Other_details = []


for link in URL:
    
    driver.get(link)
    time.sleep(3)
    try:
        ret = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[20]/div[2]/div/div/div/div[1]/div[2]/a')
        exd = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[12]/div/div/b')
        ava = driver.find_element_by_id('availability')
        time.sleep(1)
        oth = driver.find_element_by_xpath('//ul[@class="a-unordered-list a-vertical a-spacing-mini"]')
        nrat = driver.find_element_by_id('acrCustomerReviewText')
        rat = driver.find_element_by_xpath('//span[@class="a-size-medium a-color-base"]')
        time.sleep(1)
        pri = driver.find_element_by_id('priceblock_ourprice')
        tit = driver.find_element_by_id('productTitle')
        br = driver.find_element_by_id('bylineInfo')
        time.sleep(1)
        Product_title.append(tit.text)
        Brand.append(br.text.replace('Brand:','').replace('Visit the ','').replace('Store',''))
        No_ratings.append(nrat.text.replace('ratings',''))
        time.sleep(1)
        Rating.append(rat.text)
        Price.append(pri.text.replace('₹',''))
        Return.append(ret.text)
        Expected_delivery.append(exd.text)
        Availability.append(ava.text)
        Other_details.append(oth.text)
    except NoSuchElementException:
        Brand.append('-')
        Product_title.append('-')
        No_ratings.append('-')
        Rating.append('-')
        Price.append('-')
        Return.append('-')
        Expected_delivery.append('-')
        Availability.append('-')
        Other_details.append('-')
    time.sleep(2)
```

```python
guitars = pd.DataFrame({})
guitars['Brand'] = Brand
guitars['Product_title'] = Product_title
guitars['No_ratings'] = No_ratings
guitars['Rating'] = Rating
guitars['Price'] = Price
guitars['Return'] = Return
guitars['Expected_delivery'] = Expected_delivery
guitars['Availability'] = Availability
guitars['Other_details'] = Other_details
guitars['url'] = URL
guitars
```

```python
driver = webdriver.Chrome("chromedriver.exe")
driver.get('https://www.youtube.com')
time.sleep(2)
driver.find_element_by_id('search').send_keys('cristiano ronaldo')
driver.find_element_by_id('search-icon-legacy').click()
time.sleep(3)
driver.find_element_by_xpath('//a[@class="yt-simple-endpoint style-scope ytd-video-renderer"]').click()
time.sleep(4)
driver.find_element_by_xpath('//button[@class="ytp-play-button ytp-button"]').click()
```

```python
for _ in range(0,500):
    driver.execute_script('window.scrollBy(0,5000)')
    time.sleep(3)
    driver.execute_script('window.scrollBy(0,-2000)')
    time.sleep(1)
```

```python
comments = []
upvotes = []
posted_on = []

com = driver.find_elements_by_xpath('//*[@id="content-text"]')
up = driver.find_elements_by_xpath('//*[@id="vote-count-middle"]')
post = driver.find_elements_by_xpath('//*[@id="header-author"]/yt-formatted-string')

for i in com:
    try:
        comments.append(i.text)
    except NoSuchElementException:
        comments.append('NA')
        
for j in up:
    try:
        upvotes.append(j.text)
    except NoSuchElementException:
        upvotes.append('NA')
        
for k in post:
    try:
        posted_on.append(k.text)
    except NoSuchElementException:
        posted_on.append('NA')
```

    from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name,
    distance from city centre, ratings, total reviews, overall reviews, privates from price, 
    dorms from price, facilities and property description.  

```python
driver = webdriver.Chrome("chromedriver.exe")
#webpage
driver.get('https://www.hostelworld.com/')
time.sleep(4)
driver.find_element_by_id('location-text-input-field').click()
time.sleep(2)
driver.find_element_by_xpath('//*[@id="search-input-field"]').send_keys('London')
time.sleep(4)
driver.find_element_by_xpath('//*[@id="predicted-search-results"]/li[2]').click()
driver.find_element_by_id('search-button').click()
```

```python
privates_from_price = []
dorms_from_price = []
distance_city_centre = [] 

for _ in range(0,3):
    try:
        dist = driver.find_elements_by_xpath('//div[@class="subtitle body-3"]//a//span')
        priv = driver.find_elements_by_xpath('//div[@class="prices-col"]//a//div[1]//p')
        dorm = driver.find_elements_by_xpath('//div[@class="prices-col"]//a[1]//div[2]')

        for i in dist:
            distance_city_centre.append(i.text)
        
        for j in priv:
            privates_from_price.append(j.text.replace('Privates From\n',''))
        
        for k in dorm:
            dorms_from_price.append(k.text.replace('Dorms From\n',''))
        
        driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]').click()
        time.sleep(4)
    except NoSuchElementException:
        distance_city_centre.append('-')
        privates_from_price.append('-')
        dorms_from_price.append('-')
```
